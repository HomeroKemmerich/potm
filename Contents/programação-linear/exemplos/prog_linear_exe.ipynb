```python
# exe 7i
# Cria um modelo de LP - LpMinimize (default) ou LpMaximize
Lp_prob = p.LpProblem('Exercicio_7i', p.LpMaximize)

# Cria as variáveis do modelo
x1 = p.LpVariable("x1", lowBound = 0) # Create a variable x1 >= 0
x2 = p.LpVariable("x2", lowBound = 0) # Create a variable x2 >= 0
x3 = p.LpVariable("x3", lowBound = 0) # Create a variable x3 >= 0

# Função objetivo
Lp_prob += 4.56 * x1 + 3.50 * x2 + 4.60 * x3

# Restrições
Lp_prob += 1.10 * x1 + 0.45 * x2 + 1.20 * x3 <= 748
Lp_prob += 14 * x1 + 11 * x2 + 10 * x3 <= 9000
Lp_prob += x1 <= 500
Lp_prob += x2 >= 60
Lp_prob += x3 <= 100

# Mostra o modelo
print(Lp_prob)
status = Lp_prob.solve() # Solver
print(p.LpStatus[status]) # O status da solução

# Mostra a solução final do modelo
print(p.value(x1), p.value(x2), p.value(x3), '\n', p.value(Lp_prob.objective))
```

```python
# exe 7k
# Cria um modelo de LP - LpMinimize (default) ou LpMaximize
Lp_prob = p.LpProblem('Exercicio_7k', p.LpMaximize)

# Cria as variáveis do modelo
x1 = p.LpVariable("x1", lowBound = 0) # Create a variable x1 >= 0
x2 = p.LpVariable("x2", lowBound = 0) # Create a variable x2 >= 0
x3 = p.LpVariable("x3", lowBound = 0) # Create a variable x3 >= 0
x4 = p.LpVariable("x4", lowBound = 0) # Create a variable x4 >= 0

# Função objetivo
Lp_prob += 140 * x1 + 80 * x2 + 75 * x3 + 160 * x4

# Restrições
Lp_prob += x1 + x2 + x3 + x4 <= 2000
Lp_prob += x1 + x2 >= 800
Lp_prob += x3 >= 400
Lp_prob += x3 + x4 <= 1000

print(Lp_prob) # Mostra o modelo
status = Lp_prob.solve() # Solver
print(p.LpStatus[status]) # O status da solução

# Mostra a solução final do modelo
print(p.value(x1), p.value(x2), p.value(x3), p.value(x4),
      '\n', p.value(Lp_prob.objective))
```
